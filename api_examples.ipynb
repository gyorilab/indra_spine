## INDRA SPINE REST API

```python
import requests
```

```python
url = 'http://34.230.33.149:7700'
```

### Getting a subnetwork

- The `pubmed_fatigue` context is available by default
- The `classes` argument can be ommitted in which case the whole network is returned, or it can have 2 or more specific entries to get a subnetwork between those classes
- The return value is a CX network JSON (see https://home.ndexbio.org/data-model/)

```python
res = requests.post(url + '/get_network',
                    json={'name_search': 'pubmed_fatigue', 'classes': [{'name': 'Pons'}, {'name': 'Tongue'}]})
res.json()
```

### Choosing a new context

- The `build_graph` endpoint allows choosing a new context via a set of search terms (e.g., tremor) and giving it a name (e.g., `pubmed_tremor`)
- The service finds relevant PubMed IDs for the terms and takes a subset of its knowledgebase that comes from those PubMed IDs
- The new context can then be queried using the `get_network` endpoint

```python
requests.post(url + '/build_graph',
              json={'databases': ['pubmed'], 'terms': ['tremor'], 'name_search': 'pubmed_tremor'})
```

```python
res = requests.post(url + '/get_network', json={'name_search': 'pubmed_tremor'})
```
